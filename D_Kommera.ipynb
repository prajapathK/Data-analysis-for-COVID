```python
import numpy as np


import pandas as pd
dat=pd.read_excel('owid-covid-data.xlsx')
#dropping NaN files and replacing it with 0
cleandat=dat.dropna
zero_na= dat.fillna(0)
#zero_na

```

```python
#data of new deaths in japan
import matplotlib.pyplot
jp_dat=zero_na.loc[zero_na.location=='Japan',:]

new_deaths_jp=jp_dat['new_deaths']
#using length of the list as no of days  and plotting
x=np.arange(len(jp_dat))

plt.plot(x,new_deaths_jp)
plt.xlabel("days")
plt.ylabel("deaths")
```

```python
import matplotlib.pyplot as plt
#plotting
per_week=np.arange(0,len(jp_dat),7)
avg_death=np.array([])
for i in per_week:
    val=np.array((np.sum(new_deaths_jp[i:i+7])/2))
    avg_death=np.append(avg_death,val)
plt.plot(per_week,avg_death)
plt.xlabel("days")
plt.ylabel("deaths")
```
